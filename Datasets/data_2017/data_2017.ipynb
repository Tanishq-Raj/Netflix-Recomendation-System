```python
# As release date is of object type so we have to convert it into datetime format
meta['release_date']=pd.to_datetime(meta['release_date'],errors='coerce')

meta.sample(2)
```

```python
#  extracting years from release date column

meta['year']=meta['release_date'].dt.year
meta.sample(2)
```

```python
meta_2017 = meta[meta['year']==2017]

new_meta = meta_2017[['genres','id','title','year']]

new_meta.sample(6)
```

```python
new_meta.dtypes
```

```python
new_meta['id']=new_meta['id'].astype('int')
```

```python
credit = pd.read_csv(r'D:\Data Science Pojects\Netflix Recommendation System\data_2017\credits.csv')
credit.sample(2)
```

```python
data = pd.merge(new_meta,credit,on='id')
data.sample(2)
```

```python
# we will convert the "genre", "cast" and "crew" column into a list. if we observe it carefully, we will realised that it's a list containing dictionary. 
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))  #the "literal_eval" helps to convert the string into a list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.sample(2)
```

```python
# extract the individual genre from the data 
def make_genre_list(x):

    gen = []
    for i in x:
        if i.get('name')=='Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.nan
    else:
        return " ".join(gen)


```

```python
data['genres_list']=data['genres'].map(lambda x : make_genre_list(x))
data['genres_list']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.nan
    else:
        return (casts[0])
```

```python

#let's apply it on the cast column for the first actor

data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.nan
    else:
        return (casts[1])
```

```python

#let's apply it on the cast column for the first actor

data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python

def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.nan
    else:
        return (casts[2])
```

```python

data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.nan
    else:
        return (st.join(dt))
```

```python

data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
movie = data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]

movie.sample(2)
```

```python
movie.isna().sum()
```

```python
movie.dropna(inplace=True)
```

```python
#  rename the columns so that we can append it with the old 2016 movies dataset

movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
```

```python
movie.sample(2)
```

```python
movie['movie_title']= movie['movie_title'].str.lower()
```

```python
#we will be using this information later on in the tfidvectorizor
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
movie
```
